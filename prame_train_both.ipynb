```python
MAX_GAME_LEN:int = 100
N_MATCHS:int = 1000
IS_SHOW_GAME:bool = False
IS_UPDATE:bool = True

BLACK_EXPLORE_RATE:float = 0.5
BLACK_EXPLORE_RATE_DECREE:float = BLACK_EXPLORE_RATE/N_MATCHS
BLACK_ALPHA:float = 0.9
BLACK_GAMMA:float = 0.5
BLACK_WIN_SCORE:int = 80
BLACK_DRAW_SCORE:int = -10
BLACK_LOSE_SCORE:int = -500

WHITE_EXPLORE_RATE:float = 0
WHITE_EXPLORE_RATE_DECREE:float = WHITE_EXPLORE_RATE/N_MATCHS
WHITE_ALPHA:float = 0.9
WHITE_GAMMA:float = 0.5
WHITE_WIN_SCORE:int = 80
WHITE_DRAW_SCORE:int = -10
WHITE_LOSE_SCORE:int = -500

n_wins, n_draws, n_losses = 0, 0, 0
for i in range(N_MATCHS):
    if IS_SHOW_GAME:
        print('game', i)
    ch = Checkers()
    black_explore_rate:float = BLACK_EXPLORE_RATE - (i)*BLACK_EXPLORE_RATE_DECREE
    black_player = BlackCheckerAgent(    
        weight_path = "src/prame_q_learn/agent/weight/black_weight_2.json",        
        explore_rate= black_explore_rate,
        alpha= BLACK_ALPHA,
        gamma= BLACK_GAMMA,
        win_score= BLACK_WIN_SCORE,
        draw_score= BLACK_DRAW_SCORE,
        lose_score= BLACK_LOSE_SCORE,
                                    )
    
    # black_player = MinimaxPlayer(
    #     'black',
    #     value_func=partial(first_order_adv, 'black', 200, 100, 20, 0),
    #     # The provided legal moves might be ordered differently
    #     rollout_order_gen=lambda x: sorted(x),
    #     search_depth=4,
    #     seed=i)
    

    white_explore_rate:float = WHITE_EXPLORE_RATE - (i)*WHITE_EXPLORE_RATE_DECREE
    white_player = WhiteCheckerAgent(weight_path = "src/prame_q_learn/agent/weight/white_weight_2.json",        
                                    explore_rate= white_explore_rate,
                                    alpha= WHITE_ALPHA,
                                    gamma= WHITE_GAMMA,
                                    win_score= WHITE_WIN_SCORE,
                                    draw_score= WHITE_DRAW_SCORE,
                                    lose_score= WHITE_LOSE_SCORE,
                                    )

    winner = play_a_game(ch, black_player.next_move, white_player.next_move, MAX_GAME_LEN,is_show_detail = IS_SHOW_GAME)

    # Play with a minimax player
    # play_a_game(ch, keyboard_player_move, white_player.next_move)
    # Keep scores
    white_result:RESULT_TYPE
    black_result:RESULT_TYPE
    message:str = ""
    if winner == 'black':
        n_wins += 1
        white_result = RESULT_TYPE.LOSE
        black_result = RESULT_TYPE.WIN
        message = "black win"
    elif winner is None:
        n_draws += 1
        white_result = RESULT_TYPE.DRAW
        black_result = RESULT_TYPE.DRAW
        message = "draw"
    else:
        n_losses += 1
        white_result = RESULT_TYPE.WIN
        black_result = RESULT_TYPE.LOSE
        message = "white win"
    if IS_UPDATE:
        white_player.get_result(white_result)
        black_player.get_result(black_result)
    print(f"round : {i+1}/{N_MATCHS} {message}")

print('black win', n_wins, 'draw', n_draws, 'loss', n_losses)
```
